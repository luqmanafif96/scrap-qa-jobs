```python
import pandas as pd
import os
```

#### Write the path to fetch data

```python

path = "./data"
files = [file for file in os.listdir(path) if not file.startswith('.')]

all_fill_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_fill_data = pd.concat([all_fill_data, current_data])
    
all_fill_data.to_csv("all_jira_copy.csv", index=False)
```
